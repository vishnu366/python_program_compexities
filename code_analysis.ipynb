```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = './final_merged_github_analysis.csv'  # Add your file path here
data = pd.read_csv(file_path)

# Display initial dataset information
print("Initial data information:")
print(data.info())

# Drop rows with NA values
cleaned_data = data.dropna(subset=['Radon MI (Numeric)'])

# Check for any remaining NA values
print("\nRemaining NA values after cleaning:")
print(cleaned_data.isnull().sum())

# Display the first few rows of the cleaned data
cleaned_data.head()
```

Plot 1: Distribution of Maintainability Index (MI)

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Project', y='Radon MI (Numeric)', data=cleaned_data)
plt.title('Maintainability Index (MI) for Each Project')
plt.xticks(rotation=90)  # Rotate project names for better readability
plt.xlabel('Project')
plt.ylabel('Maintainability Index (MI)')
plt.show()
```

Plot 2: Size of Project vs Maintainability Index (MI)

```python
# Group data by size and compute the average MI for each size category
size_grouped = cleaned_data.groupby('size_in_kb')['Radon MI (Numeric)'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='size_in_kb', y='Radon MI (Numeric)', data=size_grouped)
plt.title('Average Maintainability Index by Project Size (KB)')
plt.xlabel('Project Size (KB)')
plt.ylabel('Average Maintainability Index (MI)')
plt.show()
```

Plot 3: Category vs Maintainability Index (MI)

```python
plt.figure(figsize=(10,6))
sns.barplot(x='category', y='Radon MI (Numeric)', data=cleaned_data)
plt.title('Project Category vs Maintainability Index (MI)')
plt.xlabel('Project Category (Small, Medium, Large)')
plt.ylabel('Average Maintainability Index (MI)')
plt.show()
```

Plot 4: Cyclomatic Complexity vs Maintainability Index (MI)

```python
# Group data by cyclomatic complexity and compute the average MI
complexity_grouped = cleaned_data.groupby('Radon Cyclomatic Complexity')['Radon MI (Numeric)'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Radon Cyclomatic Complexity', y='Radon MI (Numeric)', data=complexity_grouped)
plt.title('Cyclomatic Complexity vs Maintainability Index')
plt.xlabel('Cyclomatic Complexity')
plt.ylabel('Average Maintainability Index (MI)')
plt.show()
```

Bar Plot 5: Number of Functions vs Maintainability Index (MI)

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Number of Functions', y='Radon MI (Numeric)', data=cleaned_data)
plt.title('Number of Functions (Cyclomatic Complexity) vs Maintainability Index (MI)')
plt.xlabel('Number of Functions (Cyclomatic Complexity)')
plt.ylabel('Maintainability Index (MI)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming cleaned_data is the DataFrame after cleaning

# Set the figure size
plt.figure(figsize=(12, 8))

# Melt the data to create a "tidy" format for seaborn plotting
melted_data = cleaned_data.melt(id_vars=['category', 'Radon MI (Numeric)'], 
                                value_vars=['Lines of Code', 'Number of Functions'], 
                                var_name='Metric', value_name='Value')

# Create a grouped bar plot
sns.barplot(x='category', y='Value', hue='Metric', data=melted_data, palette="Set2")

# Add the MI (Maintainability Index) as a line plot on the secondary axis
ax2 = plt.gca().twinx()
sns.pointplot(x='category', y='Radon MI (Numeric)', data=cleaned_data, ax=ax2, color='black')

# Set the plot titles and labels
plt.title('Comparison of Lines of Code, Number of Functions, and MI across Categories')
plt.xlabel('Category')
plt.ylabel('Lines of Code / Number of Functions')

ax2.set_ylabel('Maintainability Index (MI)')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a scatter plot for Lines of Code vs MI
sns.scatterplot(x='Lines of Code', y='Radon MI (Numeric)', data=cleaned_data, color='blue', label='Lines of Code', s=100)

# Create a scatter plot for Number of Functions vs MI
sns.scatterplot(x='Number of Functions', y='Radon MI (Numeric)', data=cleaned_data, color='green', label='Number of Functions', s=100)

# Set plot titles and labels
plt.title('Lines of Code / Number of Functions vs Maintainability Index (MI)')
plt.xlabel('Lines of Code / Number of Functions')
plt.ylabel('Maintainability Index (MI)')

# Add a legend
plt.legend(loc='upper right')

# Display the plot
plt.tight_layout()
plt.show()
```
